## Bibliotecas

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## Fundamentals

```python
G = nx.Graph()              # For undirected graph
# G = nx.DiGraph()          # For directed grphs
# G = nx.MultiGraph()       # A graph with multiple edges between nodes
# G = nx.MultiDiGraph()     # Multigraph that is also directed
```

```python
G.add_edge(1,2)             # by creating the edge I create the nodes
G.add_edge(2,3, weight=0.9) # we can define a weight
```

```python
G.add_edge("A","B")         # We can use letters
G.add_edge("B", "B")        # We can make a loop
G.add_node("C")             # We can add nodes without connecting it
G.add_node(print)           # We can pass anything that is an object as a node
```

```python
# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
# We can create a graph from an edge list
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]

G = nx.from_edgelist(edge_list)

# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
# We can create a graph from an edge list
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]

G = nx.Graph()
G.add_edges_from(edge_list)

# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

## Degree of Nodes

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]
G = nx.Graph()
G.add_edges_from(edge_list)

# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
print(dict(G.degree)[2])
print(dict(G.degree)[3])
```

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]
G = nx.DiGraph()
G.add_edges_from(edge_list)

# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
print(dict(G.in_degree)[3])
print(dict(G.out_degree)[3])
print(dict(G.degree)[3])
```

## Centrality

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7),
            (2, 8), (8, 9), (9, 4)]
G = nx.Graph()
G.add_edges_from(edge_list)

# Visualizing the graph
nx.draw_planar(G, with_labels=True)
plt.show()
```

```python
print(nx.degree_centrality(G))
print(nx.closeness_centrality(G))
print(nx.eigenvector_centrality(G))
print(nx.betweenness_centrality(G))
```

```python
G1 = nx.complete_graph(5)
G2 = nx.complete_graph(5)
G2 = nx.relabel_nodes(G2, {0 : "A", 1: "B", 2: "C", 3: "D", 4: "E"})
G_connector = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([G1, G2, G_connector])

# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
print(nx.degree_centrality(G))
print(nx.betweenness_centrality(G))
```

## Conected Components

```python
G1 = nx.complete_graph(5)
G2 = nx.complete_graph(5)
G2 = nx.relabel_nodes(G2, {0 : "A", 1: "B", 2: "C", 3: "D", 4: "E"})
G_connector = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([G1, G2, G_connector])

# Visualizing the graph
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
print(list(nx.connected_components(G)))
```
